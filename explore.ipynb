```python
import fitz
import csv

header = [
 'Sl. No',
 'Merit',
 'Adm. Roll No std_name',
 'Gender',
 'Student ID',
 'Mobile',
 'Email'
]

all_students = []

doc = fitz.open("data/raw/WMBA Summer 2022 Admitted Student List and Student ID.pdf")
for page_num in range(doc.page_count):
    page = doc.load_page(page_num)
    info = page.get_text().split("\n")
    info = info[8:]
    rows_of_student = [ [info[student_sl], 
                         info[student_sl+1], 
                         info[student_sl+2],
                         info[student_sl+3],
                         info[student_sl+4],
                         info[student_sl+5],
                         info[student_sl+6],
                         info[student_sl+7],] 
                       for student_sl in range(0,len(info)-1,8) ]
    all_students += rows_of_student
print(all_students)
```

```python
with open('data/processed/admitted_student_info.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(all_students)
```
